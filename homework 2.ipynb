```python
cnames = cnames.drop_duplicates() #In here we overwrite the variable using drop_duplicates bc we can obtain a Series without the duplicates
cnames
```

**7.2.** Print the names that do not represent countries.

```python
# Since the top rows represent aggregates of countries (e.g., world regions), we can identify the point where the list transitions from regions to individual countries.
# Interestingly, both groups (regions and countries) are alphabetically ordered. The aggregates are alphabetically sorted until 'World', 
# and then the countries begin with 'Afghanistan' in alphabetical order.
# We could start the filter from 'World' or its index, but here we will implement an automatic detection solution that finds the transition point.
# Later, we can verify the exact index of the start_countries.

# Detect the transition in alphabetical order using iloc
for i in range(1, len(cnames)):
    if cnames.iloc[i].lower() < cnames.iloc[i-1].lower():  # In Python, we can compare strings using the < > operators based on alphabetical order
        # We'll take advantage of this to find where the alphabetical order breaks. 
        # A key step is to normalize the strings inside the loop, converting everything to lowercase,
        # because comparing lowercase and uppercase letters can cause errors.
        # In Python, lowercase letters have higher Unicode values than uppercase letters, so ignoring case is important.
        start_countries = i
        break

print(cnames[0:start_countries])  # start_paises = 49, meaning the list ends with "World"
```

You can take advantage of the structure of the dataset to realize that aggregates (Continents, Regions, etc) are all located on the top of the series 'cnames'. Moreover, since the series is small you can easily validate this assumption manually and then use that information to extract a slice of all the entries that represent non-countries entities.<br/>

**8.** In the next cell filter out, from **wdi**, the rows in which 'Country Name' represents an aggregate of countries.<br/>

```python
not_countries = cnames[0:start_countries].tolist()  # Using the previously defined start_countries index, we create a list to filter out the aggregates
wdi = wdi[~wdi["Country Name"].isin(not_countries)]  # Using ~ (negation) ensures that wdi["Country Name"] does not contain any of the values in the list
# isin() method checks if the column values match any in the list.
```

```python
#1

# 'Country Code' is the index, and we are going to use it to represent the countries that satisfy the following condition
# We are selecting countries with a population between 9 million and 11 million (10 million +/- 1 million). 
#The & operator ensures that both conditions are true, creating a closed interval
pop_10M = data[(data['Population'] >= 9000000) & (data['Population'] <= 11000000)].copy() #making this a copy to not have problems in the original data

# Display only the 'Population' column for these countries
pop_10M[['Population']]
```

```python
#4
#so first we need to define a function to classify each country based on the parameter population:
def classify_country(population):
    if population < 1000000: #Tiny - population < 1 000 000
        return 'Tiny'
    elif population < 5000000: #Very Small - 1 000 000 <= population < 5 000 000, we just use the upper bound beacause if it was less than 1 000 000 (not included),
        #the funtion would stop in the first line
        return 'Very Small'
    elif population < 15000000: #Small - 5 000 000 <= population < 15 000 000
        return 'Small'
    elif population < 30000000: #Medium - 15 000 000 <= population < 30 000 000
        return 'Medium'
    elif population < 100000000: #Large - 30 000 000 <= population < 100 000 000
        return 'Large'
    else:
        return 'Huge' #Huge - 100 000 000 <= population

data['Country classification'] = data['Population'].apply(classify_country) #we create a new column applying the funtion to the column population
data['GDP per capita']= data['GDP']/data['Population'] #and since we only have the gdp, we can just use column GDP and divide by the population to obtain the gdp per capita 
 
pd.DataFrame(data.groupby('Country classification')['GDP per capita'].agg(['mean', 'std'])) #When grouping by the column "Country classification" (unique values),
#we aggregate the GDP per capita for each value by calculating the mean and the standard deviation.
```

```python
#5
# Function to create a binary series based on whether the values are above the mean.
# Inspired by: pds_lab4_24 
def above_mean_binarizer(df, col):
    """
    Creates a binary series based on whether the values in a given column are above or below the mean (1 if above mean, 0 otherwise, NaN if NaN)
    It preserves NaN values as NaN in float format.
    """
    mean_value = df[col].mean()  # as a starting point we calculate the mean of the input column

    # Then we create the series, preservating the Nan's and giving one if it is above average and 0 otherwise
    binary_series = [
        np.nan if pd.isna(value) else 1 if value > mean_value else 0
        for value in df[col]
    ]
    
    # Converting to pd.Series with dtype=object to ensure we preserve NaN as NaN and retain integers (1 and 0) instead of floats (1.0 and 0.0).
    # Previously, we had some problems as this function returned floats due to NaN handling, and this approach resolves that issue. 
    return pd.Series(binary_series, index=df.index, dtype=object)

#just a test for the Nan values
result = above_mean_binarizer(pd.DataFrame({"test": [1, 2, 3, None, 5]}), "test")
print(f'Test to see if None is converted to NaN:\n{result}\nNone converted to NaN: {pd.isna(result[3])}')

#actual column
print(f'\nTest in the column hc \n{above_mean_binarizer(data, 'hc')}')
```

```python
#6
#What is the average GDP per capita of the countries after being grouped by size classification and whether the human capital was above or below average?
#Hint: as an example, two of the groups should be (1) tiny and human capital below average, (2) tiny and human capital above average.
#we create a column with the function to obtain the series and then...
data['hc_binary'] = above_mean_binarizer(data, 'hc')
data.groupby(['Country classification', 'hc_binary'])['GDP per capita'].mean() #we groupt by Country classification and hc_binary, calculating the average GDP per capita
#for each combination
```

```python
#7
#What is the average GDP per capita of the countries after being grouped by whether the human capital was above or below average
#and whether the gini coefficient was above or below average?

data['gini_binary'] = above_mean_binarizer(data, 'GINI') #created the needed column with the function
data.groupby(['hc_binary', 'gini_binary'])['GDP per capita'].mean() #same logic as before but now grouping 'hc_binary' and 'gini_binary' to obatain the average GDP per capita
#of these combinations
```

```python
#8
#What is the name of the country that has the highest GDP per capita (the max GDP per capita betw the countries that satisfied the other conditions),
#a Gini coefficient below average (==0) and a level of human capital below average(==0)?
result_8 = data[data['GDP per capita'] == data[(data['gini_binary'] == 0) & (data['hc_binary'] == 0)]['GDP per capita'].max()].index[0]
print(f"Country with the highest GDP per capita, Gini coefficient below average, and human capital below average is: \n "
      f"{wdi[wdi['Country Code'] == result_8]['Country Name'].drop_duplicates().to_list()[0]} ({result_8})") #get the actually 'name of the country' as specified
```

```python
#9
#name of the country that has the highest GDP per capita, a Gini coefficient below average for its size classification,
#and a level of human capital below average for its size classification?

#This question is interpreted as, btw the countries with a Gini coefficient below average for its size classification 
#-meaning comparing to its size classification, NOT the complete column column-
#and a level of human capital below average for its size classification, what is the one with the highest GDP per capita, so...
valid_country_indexes = []

mean_values_by_classification = data.groupby("Country classification")[["GINI", "hc"]].mean() #we group by Country classification to obtain the mean for GINI & hc for each classification
#and then, we can compare the values of GINI and hc of each country with the ones of their classification:
for country_code, row in data.iterrows():
    if (row["GINI"] < mean_values_by_classification.loc[row["Country classification"], "GINI"]) and \
       (row["hc"] < mean_values_by_classification.loc[row["Country classification"], "hc"]):
        valid_country_indexes.append(country_code) #we obtain all the countries codes that satisfied this condition to then filter data with it 

result_9 = data.loc[valid_country_indexes, "GDP per capita"].idxmax() #and finally we can get the max GDP per capita
print(
    f"Country with the highest GDP per capita, Gini coefficient below average for its size classification, "
    f"and human capital below average for its size classification is: \n"
    f"{wdi[wdi['Country Code'] == result_9]['Country Name'].drop_duplicates().values[0]} ({result_9})") #get the actually 'name of the country' as specified

#To check: display mean_values_by_classification and data.loc[valid_country_indexes]
```

```python
#10 
#What is the name of the country that has the largest % increase in GDP between 1980 and 2010? HINT: You will need to use the wdi dataframe.

#We got the 'Indicator Code' of th GDP from uestion 14 again to filter the df wdi
wdi_gdp = wdi[wdi['Indicator Code'] == 'NY.GDP.MKTP.CD'].copy() #copy just to be safe

wdi_gdp_sample = wdi_gdp.loc[:, ["Country Code", "1980", "2010"]].dropna() #drop nan so we can safly perform the calculations
wdi_gdp_sample["% gdp increase"] = (wdi_gdp_sample["2010"] - wdi_gdp_sample["1980"]) / wdi_gdp_sample["1980"] * 100 #% increase = (New Number - Original Number) √∑ Original Number √ó 100
result10 = wdi_gdp_sample[wdi_gdp_sample['% gdp increase'] == wdi_gdp_sample['% gdp increase'].max()]

print(f"Country that has the largest % increase in GDP between 1980 and 2010 is: \n"
      f"{wdi[wdi['Country Code'].isin(result10['Country Code'])]['Country Name'].values[0]} ({result10.iloc[0, 0]}) "
      f"with a {result10.iloc[0, 3]:.2f}% increase")
```
